```python
import requests
import json
import pandas as pd

# Leer el archivo Excel con la columna entityId
# Cambia esta ruta por la de tu archivo real si hace falta
# El Excel debe contener una columna llamada entityId

df = pd.read_excel("demo_hostids.xlsx")
lista = df["entityId"].tolist()
i = 0
for j in lista:
    HostID = str(lista[i])
    api_token = "abc"
    URL = "https://ejemplo.live.dynatrace.com/api/v2/settings/objects?validateOnly=false"

    header = {
        'accept': 'application/json; charset=utf-8',
        'Authorization': 'Api-Token ' + str(api_token),
        'Content-Type': 'application/json; charset=utf-8'
    }

    dataOA = '[{"schemaId":"builtin:host.monitoring","schemaVersion":"1.3","scope":"' + HostID + '","value":{"enabled":false,"autoInjection":true}}]'

    params = {
        "format": "json"
    }

    response = requests.post(url=URL, headers=header, data=dataOA)
    print(response.text)
    i += 1
```
